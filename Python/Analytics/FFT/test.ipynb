```python
import scipy
import scipy.fftpack
import numpy
import matplotlib.pyplot as plt
#plt.ion()

filename = 'C:\\Users\\Ranojay.Sen\\AppData\\Roaming\\Schlumberger\\Techlog\\PythonScripts\\panel.txt'         ## name of file in which the waveforms are saved
spacing = 0.5                    ## distance between receivers, in feet
receivers = 8                    ## Number of receivers on the tool
tr = 12                           ## Transmitter-to-receiver distance, in feet

gate = 20                       ## Full width of the semblance gate, in digi units

v_min = 3000                     ## for display purposes only, minimum velocity in ft/s
v_max = 14000                    ## for display purposes only, max velocity in ft/s
t_min = 0
t_max = 7000
fmax = 14

pi = numpy.arccos(-1)
ln2 = numpy.log(2.0)
print(ln2)
# Filter paramters

filt_type = "minimum"
low6db = 4
lowpass = 4.3
hipass = 10
hi6db = 10.3
LowSlope = 38
HiSlope = 48

# Read waveforms in from the data file.  

length = 2048
infile = open(filename)
panel = numpy.zeros((receivers,length))
time = numpy.zeros(length)
i = 0
for line in infile:
        L = line.split()
        time[i] = float(L[0])
        for j in range(receivers):
                panel[j,i] = float(L[j+1])
        i += 1                
infile.close()

digi = time[1] - time[0]          # Defines the sampling rate of the tool, in microseconds
length = min(length,i)
tvack = int((t_max-t_min)/digi)
vvack = 200

## Since FFT algorithms only really work properly on data sets with length equal to a power of two, find the first power of two that's
## larger than the actual length of the input data.

filt_length = max(length,32)
P2 = int(numpy.log(filt_length) / ln2) + 1

filt_length = 2**P2

#filt_length = 1024
fft_length = int(filt_length/2)


# Define variables

time = numpy.zeros(length)
newpanel = numpy.zeros((receivers,filt_length))
shiftpanel = numpy.zeros((receivers,filt_length))
filtpanel = numpy.zeros((receivers,filt_length))
outpanel = numpy.zeros((length,receivers))
outpanel2 = numpy.zeros((filt_length,receivers))
stackout = numpy.zeros((tvack,receivers))
sembpanel = numpy.zeros((tvack,vvack))
freq = numpy.zeros(fft_length)
filt = numpy.zeros(filt_length)
filt_phase = numpy.zeros(filt_length)
outfilt = numpy.zeros(filt_length)
imag_filt = numpy.zeros(filt_length)
v_axis = numpy.zeros(vvack)
t_axis = numpy.zeros(tvack)
t0 = numpy.zeros(vvack)
maxvack = numpy.zeros(vvack)

for i in range(length):
        time[i] = i*digi
        for r in range(receivers):
            newpanel[r][i] = panel[r][i]

# Display waveforms, calculate FFT, and plot the transform.

plt.figure(1, figsize=(12,12))
plt.clf()
scaler = max(abs(newpanel[0][:length]))*2
for i in range(receivers):
        for j in range(length):
                outpanel[j][i] = newpanel[i][j]/scaler + (i+1)
plt.plot(time,outpanel,color="grey",linewidth=1)
            
nyquist = 1/(2*digi)
df = 1000*(nyquist/fft_length)
fftp = scipy.fft(newpanel)

# Create an XALOG-style filter wavelet

lc_sample = int(low6db/df)
lp_sample = int(lowpass/df)
hp_sample = int(hipass/df)
hc_sample = int(hi6db/df)

# set magnitude values for the low-frequency slope

if (lc_sample == lp_sample):
        s = 0.0
        fac = 1.0
else:
        s = .5*pi/float(lp_sample - lc_sample)
        fac = 0.5

join_sam = lc_sample - 1
r = 1.0/float(join_sam)
b = s - r*float(LowSlope)
if (b>0):
        a = 2.0*b
        fac = .5*pi*fac
        for i in range(1,lc_sample):
                filt[i] = fac*(1.0 + numpy.exp(a*(i-join_sam-1)))*pow(r*(i-1),float(.5*LowSlope/pi))
        if(LowSlope == 0):
                filt[0] = fac*(1.0 + numpy.exp(a*(i-join_sam-1)))
else:
        a = -b
#        print('poink!')
        for i in range(1,lc_sample):
                filt[i] = fac*(2.0 + numpy.exp(a*(i-join_sam-1)))*pow(r*(i-1),float(.5*LowSlope/pi))
        if(LowSlope == 0):
                filt[0] = fac*(2.0 + numpy.exp(a*(i-join_sam-1)))

# set magnitude values between lc_sample and lp_sample

if (lc_sample != lp_sample):
        s = .5*pi/float(lp_sample - lc_sample)
        for i in range(lc_sample, lp_sample+1):
                filt[i] = .5*pi*(1.0 + numpy.cos(s*(lp_sample - i)))

# set magnitude values for pass frequencies

for i in range(lp_sample, hp_sample+1):
        filt[i] = pi

# set magnitude values between hc_sample and hp_sample

if (hc_sample != hp_sample):
        s = .5*pi/float(hc_sample - hp_sample)
        for i in range(hp_sample, hc_sample+1):
                filt[i] = .5*pi*(1.0 + numpy.cos(s*(hp_sample - i)))

# set magnitude values for the high-frequency slope

if (hc_sample == hp_sample):
        s = 0.0
        fac = 1.0
else:
        s = -.5*pi/float(hc_sample-hp_sample)
        fac = 0.5

join_sam = hc_sample + 1
r = 1.0/float(hc_sample - 1)
b = -(s*r*float(HiSlope))
if b > 0:
        a = -2.0*b
        fac = .5*pi*fac
        for i in range(join_sam, fft_length):
                filt[i] = fac*(1.0 + numpy.exp(a*(i-hc_sample)))/pow(r*i, float(.5*HiSlope/pi))
else:
        a = b
#        print ('ping!')
        for i in range(join_sam, fft_length):
                filt[i] = fac*(2.0 + numpy.exp(a*(i-hc_sample)))/pow(r*i, float(.5*HiSlope/pi))

for i in range(fft_length):
        filt[(fft_length*2)-i-1] = filt[i]
        freq[i] = 1000*i*nyquist/fft_length
#
#   Plot FFTs and filter
#
plt.figure(2, figsize=(12,12))
plt.clf()
fftmax = 0
fsteps = int(fmax/freq[1])
for i in range(receivers):
        absfft = abs(fftp[i][:fsteps])
        plt.plot(freq[:fsteps],absfft)
        fftmax = max(fftmax,max(absfft)/pi)
plt.plot(freq[:fsteps],filt[:fsteps]*fftmax,'r',lw=2)
plt.xlabel('frequency [kHz]')
plt.title('FFTs')
#
#Change to minimum-phase filter, if desired
#
ii = scipy.sqrt(-1)
lz = scipy.log(.005)
if filt_type == 'minimum':
        for i in range(filt_length):
                if filt[i] > 0:
                        imag_filt[i] = scipy.log(filt[i])
                else:
                        imag_filt[i] = lz

        filt_phase = scipy.fftpack.hilbert(imag_filt)
        
        for i in range(filt_length):
                mag = float(filt[i])
                filt[i] = mag*scipy.cos(filt_phase[i])
                imag_filt[i] = mag*scipy.sin(filt_phase[i])

        filt = filt + ii*imag_filt
#
#Calculate and display filtered waveforms
#
plt.figure(2)
for i in range(receivers):
#        for j in range(length):
#                fftp[i][j] = fftp[i][j]*filt[j]
    fftp[i] = fftp[i]*filt
plt.plot(freq[:fsteps],abs(fftp[0][:fsteps]/pi),'k--',lw=2)

ipanel = scipy.ifft(fftp)

for i in range(receivers):
        for j in range(length):
                filtpanel[i][j] = float(ipanel[i][j])
scaler = max(abs(filtpanel[0]))*2

plt.figure(1)
for i in range(receivers):
        for j in range(length):
                outpanel[j][i] = filtpanel[i][j]/scaler + i+1
plt.plot(time,outpanel)
plt.xlabel('time [us]')
plt.title('Raw and filtered waveforms')

delta_v = (v_max-v_min)/vvack
delta_t = digi
vgate = int(gate/2)
#
#set up some parameters for plotting
#
for i in range(vvack):
        v_axis[i] = v_min + delta_v*float(i)
        t0[i] = tr / v_axis[i] * 1e6

print('Calculating VACK...')
for i in range(vvack):
        for r in range(receivers):
#       midpoint-centered moveout
                moveout = 1e6*((r-float(receivers-1)/2)*spacing)/v_axis[i]
                for j in range(tvack):
                        t = t_min + delta_t * float(j)
                        dt = t + moveout
                        shift = int(dt/digi)
                        frac = dt/digi - shift
                        if(shift >=0) and (shift+1)<length:
                                shiftpanel[r][j] = filtpanel[r][shift]*(1-frac) + filtpanel[r][shift+1]*frac
                        else:
                                shiftpanel[r][j] = 0

        numerator = denominator = 0
        for j in range(tvack):
                t = j
                if j == 0:
#                       for k in range(t,t+vgate+1):
                       for k in range(0,gate+1):
                                s = s2 = 0
                                for r in range(receivers):
                                        foo = shiftpanel[r][k]
                                        s += foo
                                        s2 += foo*foo
                                numerator += s*s
                                denominator += s2
                else:
                        s = s2 = d = d2 = 0
                        for r in range(receivers):
#                                 if t-vgate >= 0:
#                                         foo = shiftpanel[r][t-vgate]
                                 if t-1 >= 0:
                                         foo = shiftpanel[r][t-1]
                                         d += foo
                                         d2 += foo*foo
#                                 if t+vgate+1 < length:
#                                         foo = shiftpanel[r][t+vgate+1]
                                 if t+gate < length:
                                         foo = shiftpanel[r][t+gate]
                                         s += foo
                                         s2 += foo*foo
                        numerator += (s*s-d*d)
                        denominator += (s2-d2)
                if(denominator > 0):
                        cell = (numerator/denominator) / receivers
                        sembpanel[j,i] = 1000 * min(1, max(cell,0))
#                        print numerator, denominator, cell
                else:
                        sembpanel[j,i] = 0
                if(sembpanel[j,i] > maxvack[i]):
                        maxvack[i] = sembpanel[j,i]
                

plt.figure(3, figsize=(12,12))
plt.clf()
plt.imshow(sembpanel, extent = [v_min, v_max, t_max, t_min])
plt.plot(v_axis,t0,'white',lw=2)
plt.xlabel('Velocity [ft/s]')
plt.ylabel('Time [usec]')
plt.title('Semblance panel')
#
# Stack test
#
stackvel = 11380
stacktime = 1211

for r in range(receivers):
    moveout = 1e6*((r-float(receivers-1)/2)*spacing)/stackvel
    for j in range(tvack):
            t = t_min + delta_t * float(j)
            dt = t + moveout
            shift = int(dt/digi)
            frac = dt/digi - shift
            if(shift >=0) and (shift+1)<length:
                shiftpanel[r][j] = filtpanel[r][shift]*(1-frac) + filtpanel[r][shift+1]*frac
            else:
                shiftpanel[r][j] = 0
            t_axis[j] = t

plt.figure(3)
plt.plot(stackvel,stacktime,'o',color='black')

plt.figure(4, figsize=(12,12))
for r in range(receivers):
        for j in range(tvack):
                stackout[j][r] = shiftpanel[r][j]/scaler + (r+1)
plt.plot(t_axis,stackout)

#plt.plot([stacktime-float(vgate)*digi,stacktime-float(vgate)*digi],[0,receivers+.5],color='black')
#plt.plot([stacktime+float(vgate)*digi,stacktime+float(vgate)*digi],[0,receivers+.5],color='black')
plt.plot([stacktime,stacktime],[0,receivers+.5],color='black')
plt.plot([stacktime+float(gate)*digi,stacktime+float(gate)*digi],[0,receivers+.5],color='black')
plt.xlabel('Time [us]')
plt.title('Stacked waveforms')


stackvel = 6680
stacktime = 2200

for r in range(receivers):
    moveout = 1e6*((r-float(receivers-1)/2)*spacing)/stackvel
    for j in range(tvack):
            t = t_min + delta_t * float(j)
            dt = t + moveout
            shift = int(dt/digi)
            frac = dt/digi - shift
            if(shift >=0) and (shift+1)<length:
                shiftpanel[r][j] = filtpanel[r][shift]*(1-frac) + filtpanel[r][shift+1]*frac
            else:
                shiftpanel[r][j] = 0
            t_axis[j] = t

plt.figure(5, figsize=(12,12))
plt.clf()
plt.imshow(sembpanel, extent = [v_min, v_max, t_max, t_min])
plt.plot(v_axis,t0,'white',lw=2)
plt.plot(stackvel,stacktime,'o',color='black')
plt.xlabel('Velocity [ft/s]')
plt.ylabel('Time [usec]')
plt.title('Semblance panel')

plt.figure(6, figsize=(12,12))
for r in range(receivers):
        for j in range(tvack):
                stackout[j][r] = shiftpanel[r][j]/scaler + (r+1)
plt.plot(t_axis,stackout)

#plt.plot([stacktime-float(vgate)*digi,stacktime-float(vgate)*digi],[0,receivers+.5],color='black')
#plt.plot([stacktime+float(vgate)*digi,stacktime+float(vgate)*digi],[0,receivers+.5],color='black')
plt.plot([stacktime,stacktime],[0,receivers+.5],color='black')
plt.plot([stacktime+float(gate)*digi,stacktime+float(gate)*digi],[0,receivers+.5],color='black')
plt.xlabel('Time [us]')
plt.title('Stacked waveforms')

stackvel = 3830
stacktime = 3600

for r in range(receivers):
    moveout = 1e6*((r-float(receivers-1)/2)*spacing)/stackvel
    for j in range(tvack):
            t = t_min + delta_t * float(j)
            dt = t + moveout
            shift = int(dt/digi)
            frac = dt/digi - shift
            if(shift >=0) and (shift+1)<length:
                shiftpanel[r][j] = filtpanel[r][shift]*(1-frac) + filtpanel[r][shift+1]*frac
            else:
                shiftpanel[r][j] = 0
            t_axis[j] = t

plt.figure(7, figsize=(12,12))
plt.clf()
plt.imshow(sembpanel, extent = [v_min, v_max, t_max, t_min])
plt.plot(v_axis,t0,'white',lw=2)
plt.plot(stackvel,stacktime,'o',color='black')
plt.xlabel('Velocity [ft/s]')
plt.ylabel('Time [usec]')
plt.title('Semblance panel')

plt.figure(8, figsize=(12,12))
for r in range(receivers):
        for j in range(tvack):
                stackout[j][r] = shiftpanel[r][j]/scaler + (r+1)
plt.plot(t_axis,stackout)

#plt.plot([stacktime-float(vgate)*digi,stacktime-float(vgate)*digi],[0,receivers+.5],color='black')
#plt.plot([stacktime+float(vgate)*digi,stacktime+float(vgate)*digi],[0,receivers+.5],color='black')
plt.plot([stacktime,stacktime],[0,receivers+.5],color='black')
plt.plot([stacktime+float(gate)*digi,stacktime+float(gate)*digi],[0,receivers+.5],color='black')
plt.xlabel('Time [us]')
plt.title('Stacked waveforms')

plt.show()
```
