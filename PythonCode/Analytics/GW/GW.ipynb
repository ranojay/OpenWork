```python
'''
===============================
State wise deep tubewell survey
===============================

A very basic demonstration of Dataframes and plots
'''


#State wise distribution of deep water tube wells
swdistcsvfile = "..\\..\\..\\Data\\GW\\SwDeepWell\\swdistdeepwell.csv"

# import DataFrame from pandas package so that we can work with this tabular data
import pandas as pd
from pandas import DataFrame as df

#Data frame from csv file
state_bw_df = pd.read_csv(swdistcsvfile)

# now that we have a DataFrame lets peek into the data and see how it looks
state_bw_df.head()
#print(state_bw_df)
```

#Total number of records
len(state_bw_df.index)

```python
# we can also get high level information of the width and height of the data frame using shape property
state_bw_df.shape
```

```python
sum(state_bw_df["GRAND TOTAL"] )
```

```python
total_wells = sum(state_bw_df["GRAND TOTAL"] )
percent_lambda = lambda x: x/total_wells * 100 
state_bw_df['percentage'] = state_bw_df["GRAND TOTAL"].apply(percent_lambda)
state_bw_df.head()
```

```python
import matplotlib.pyplot as plt
#plt.figure(1, figsize=(12,8))
plt.pie(state_bw_df.percentage, labels=state_bw_df.STATE)
plt.title('Sate wise distribution of wells')
plt.show()
```

```python
import numpy as np
x = np.arange(state_bw_df["STATE"].count())
#plt.figure(2, figsize=(12,8))
plt.xticks(x, state_bw_df.STATE, rotation=90)
y = state_bw_df.percentage
plt.bar(x, y)
plt.show()
```

```python
#Tube wells in use
import numpy as np
x = np.arange(state_bw_df["STATE"].count())
#plt.figure(2, figsize=(12,8))
plt.xticks(x, state_bw_df.STATE, rotation=90)
y = state_bw_df["TUBEWELLS IN USE"]
plt.bar(x, y)
plt.show()
```
